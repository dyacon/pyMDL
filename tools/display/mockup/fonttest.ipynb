## Font comparison for MDL

```python
#Import PIL and use matplotlib for display
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np

#Create image
im = Image.new('1',(128,64),color=0)
d = ImageDraw.Draw(im)
```

```python
#Create fonts to compare
cherry = ImageFont.load('cherry-11-r.pil')
dfnt = ImageFont.load_default()
```

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0,outline=255)

#Display content
teststr1 = 'ABCefg 01238'
teststr2 = 'GHIJKL 56789'
teststr3 = 'abcdefghijkl'

d.text((0,0),teststr1,font=cherry,fill=255)
d.text((0,40),teststr1,font=dfnt,fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```

```python
#Create fonts to compare
arial = ImageFont.truetype('arialbd.ttf',10)
georgia = ImageFont.truetype('georgiab.ttf',10)
```

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0,outline=255)

#Display content
teststr1 = 'ABCefg 01238'
teststr2 = 'GHIJKL 56789'
teststr3 = 'abcdefghijkl'

d.text((0,0),teststr1,font=arial,fill=255)
d.text((0,20),teststr1,font=georgia,fill=255)
d.text((0,40),teststr1,font=dfnt,fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```
