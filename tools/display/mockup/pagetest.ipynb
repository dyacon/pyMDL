```python
#Define font
cherry = ImageFont.load('fonts/cherry-10-r.pil')
```

## System Display

```python
#Create a system display
systemdisp = pymdl.display.system_display()
systemdisp.font = cherry
```

```python
#Render image
img = systemdisp.renderPage()
npimg = np.asarray(img)
plt.imshow(npimg,cmap='gray')
```

```python
#Create a sensor display
tph1disp = pymdl.display.sensor_display('tph1')
tph1disp.font = cherry
```
