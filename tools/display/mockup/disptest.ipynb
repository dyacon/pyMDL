## Testing some different display configurations
Use this on Windows to preview what the display might look like on the MDL

### Create basic image

```python
#Import PIL and use matplotlib for display
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np

#Create image
im = Image.new('1',(128,64),color=0)
d = ImageDraw.Draw(im)


```

```python
#Define fonts
libm = ImageFont.truetype('LiberationMono-Regular.ttf',10)
dfnt = ImageFont.load_default()
```

### System Display

```python
dispfnt = dfnt

#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headerstr = ('MDL-700     13:45:22\n')
statstr = ('IP Address: 225.552.332.220\n'
           'DataBear V2.1: Running\n'
           'Data Trasmission: Idle')

d.multiline_text(
        (5,0),
        headerstr,
        font=dispfnt,
        fill=255,
        spacing=2)

d.multiline_text(
        (5,25),
        statstr,
        font=dispfnt,
        fill=255,
        spacing=2)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```

### Data Display

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headstr = 'TPH-1   20:05:10'
datastr = ('T:  20.56 C\n'
           'RH: 35%\n'
           'BP: 890 mb\n')

d.text((0,0),headstr,font=libm,fill=255)
d.multiline_text(
        (20,20),
        datastr,
        font=libm,
        fill=255,
        spacing=2)

#Draw a rectangle to indicate more data
d.polygon([(50,110),(50,120),(55,105)],fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```

### Test data display: default font

```python
dfnt = ImageFont.load_default()

#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headstr = ' TPH-1   20:05:10'
datastr = ('T:  20.56 C\n'
           'RH: 35%\n'
           'BP: 890 mb\n')

d.text((0,0),headstr,font=dfnt,fill=255)
d.multiline_text(
        (20,20),
        datastr,
        font=dfnt,
        fill=255,
        spacing=2)

#Draw a rectangle to indicate more data
d.polygon([(110,50),(120,50),(115,55)],fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```

#### Test system display

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headerstr = ('MDL-700     13:45:22\n'
             'Status\n')
statstr = ('DataBear: Running\n'
           'Battery:  12.35V\n'
           'No problems detected')

d.multiline_text(
        (5,0),
        headerstr,
        font=dfnt,
        fill=255,
        spacing=2)

d.multiline_text(
        (5,25),
        statstr,
        font=dfnt,
        fill=255,
        spacing=2)
d.line([(5,24),(40,24)],width=1,fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```
