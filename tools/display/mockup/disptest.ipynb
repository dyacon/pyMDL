## Testing some different display configurations
Use this on Windows to preview what the display might look like on the MDL

### Create basic image

```python
#Import PIL and use matplotlib for display
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create image
im = Image.new('1',(128,64),color=0)
d = ImageDraw.Draw(im)
```

```python
#Define font
testfont = ImageFont.load('fonts/cherry-11-r.pil')
```

### System Display

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headerstr = ('MDL-700     13:45:22\n')
statstr = ('IP: 225.552.332.220\n'
           'Status:\n'
           'DataBear V2.2 Active \n')

d.multiline_text(
        (0,0),
        headerstr,
        font=testfont,
        fill=255,
        spacing=0)

d.multiline_text(
        (0,25),
        statstr,
        font=testfont,
        fill=255,
        spacing=0)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```

### Data Display

```python
#Clear image
d.rectangle([(0,0),(128,64)],fill=0)

#Display content
headstr = 'TPH_roof1   20:05:10'
datastr = ('air_t: 20.56 C\n'
           'relat: 35.33 %\n'
           'barom: 890.22 mb\n')

d.text((0,0),headstr,font=testfont,fill=255)
d.multiline_text(
        (10,20),
        datastr,
        font=testfont,
        fill=255,
        spacing=2)

#Draw a rectangle to indicate more data
d.polygon([(115,50),(125,50),(120,55)],fill=255)

npim = np.asarray(im)
plt.imshow(npim,cmap='gray')
```
